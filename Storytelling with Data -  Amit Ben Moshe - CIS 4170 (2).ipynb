```python
import altair as alt
import pandas as pd

# Load data
music_df = pd.read_csv('Music_Mental_Health.csv')

# Define selection
selection = alt.selection_multi(fields=['Music effects'], bind='legend')

# Define chart
chart = alt.Chart(music_df).mark_bar().encode(
    x='count()',
    y='Music effects',
    color=alt.condition(selection, 'Music effects', alt.value('gray')),
    tooltip=['Music effects', 'count()']
).properties(
    width=600,
    height=400,
    title='Effects of Music on Mental Health'
).add_selection(selection)

# Add interactivity
chart.interactive()
```
